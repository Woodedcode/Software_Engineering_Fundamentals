#NESTED WHILE LOOPS

```python
i = 0

while i < 10:
  j = i
  while j < 10:
    print(j, end=" ")
    j += 1
  print()
  i += 1


###############################################
print("All the factorials with output:")
i = 10
print("Getting factorials up to", i)
while i > 0:
  factorial = 1
  j=i
  while j > 0:
    print(j, end=" ")
    factorial *= j
    j -= 1
    if j > 0:
      print('x', end=" ")
    else:
      print('=', end=" ")
  print(factorial)
  i -= 1

print("0 = 1")


print("All the factorials with output:")
i = 11
print("Getting factorials up to", i)
for i in range(i, 0, -1):
  factorial = 1
  j=i
  for j in range(i, 0 ,-1):
    print(j, end=" ")
    factorial *= j
    j -= 1
    if j > 1:
      print('x', end=" ")
    else:
      print('=', end=" ")
  print(factorial)
  i -= 1

print("0 = 1")
```
